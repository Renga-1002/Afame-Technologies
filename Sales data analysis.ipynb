### TOP CUSTOMERS TO ORDER PRODUCT MONTHLY IN CATEGORY :

```python
# Group data by CustomerID, Category, and Month
monthly_customer_category_counts = df.groupby(['Customer ID', 'Category', pd.Grouper(key='Order Date', freq='M')]).size().reset_index(name='OrderCount')
monthly_customer_category_counts
```

```python
# Get top 10 customers per category in each month
top_10_customers_per_category = monthly_customer_category_counts.groupby(['Category', 'Order Date']).apply(lambda x: x.nlargest(10, 'OrderCount')).reset_index(drop=True)
top_10_customers_per_category
```

```python
# Plotting top 10 customers' order counts per category
plt.figure(figsize=(12, 8))
for category in top_10_customers_per_category['Category'].unique():
    category_data = top_10_customers_per_category[top_10_customers_per_category['Category'] == category]
    for i, (_, row) in enumerate(category_data.iterrows(), 1):
        plt.barh(f'{category} - {row["Order Date"].strftime("%B")}', row['OrderCount'], color=plt.cm.tab10(i))
plt.xlabel('Order Count')
plt.title('Customers Order Category in Monthly')
plt.gca().invert_yaxis()  # Invert y-axis to display latest month at the top
plt.show()
```

```python
# Group data by CustomerID, Sub-Category, and Month
monthly_customer_category_counts = df.groupby(['Customer ID', 'Sub-Category', pd.Grouper(key='Order Date', freq='M')]).size().reset_index(name='OrderCount')
monthly_customer_category_counts
```

### RELATIONSHIP BETWEEN TOTAL SALES AND SUB-CATEGORY:

```python
# Aggregate data by month and subcategory
monthly_sales = df.groupby([pd.Grouper(key='Order Date', freq='M'), 'Sub-Category'])['Sales'].sum().reset_index()
```

```python
# Calculate total sales by month
total_sales = monthly_sales.groupby('Order Date')['Sales'].sum().reset_index()
```

```python
# Merge total sales with monthly sales data
merged_data = pd.merge(monthly_sales, total_sales, on='Order Date', suffixes=('_subcategory', '_total'))
```

```python
# Plot relationship between total sales and subcategories
plt.figure(figsize=(10, 6))
for subcategory in merged_data['Sub-Category'].unique():
    subset = merged_data[merged_data['Sub-Category'] == subcategory]
    plt.plot(subset['Order Date'], subset['Sales_subcategory'], marker='o', label=subcategory)
plt.plot(merged_data['Order Date'], merged_data['Sales_total'], marker='o', linestyle='--', color='black', label='Total Sales')
plt.title('Relationship between Total Sales and Subcategories')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.legend()
plt.grid(True)
plt.show()
```
